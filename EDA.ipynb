# Load Data

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

stocks = ['ANTM.JK', 'BBNI.JK', 'BBRI.JK', 'BMRI.JK', 'TLKM.JK']

for stock in stocks:
    print(f"\n\n================================== {stock} ==================================\n")
    data = yf.download(stock, start='2021-05-01', end='2025-05-01')
    data = data[['Close']].dropna()
    data.rename(columns={'Close': 'Price'}, inplace=True)

    print(data.info())
    print(data.head(5))
    print(data.tail(5))

    plt.figure(figsize=(14, 6))
    plt.plot(data.index, data['Price'], label='Closing Price', color='blue')
    plt.title('Data Set'+ stock)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# Analisis Univariat

```python
for stock in stocks:
    print(f"\n\n================================== {stock} ==================================\n")
    data = yf.download(stock, start='2021-05-01', end='2025-05-01')
    data = data[['Close']].dropna()
    data.rename(columns={'Close': 'Price'}, inplace=True)

    mean_price = round(float(data['Price'].mean()), 2)
    median_price = round(float(data['Price'].median()), 2)
    std_price = round(float(data['Price'].std()), 2)
    min_price = round(float(data['Price'].min()), 2)
    max_price = round(float(data['Price'].max()), 2)

    print("Mean Price:", mean_price)
    print("Median Price:", median_price)
    print("Min Price:", min_price)
    print("Max Price:", max_price)
    print("Std Dev:", std_price)
```

# Analisis Multivariat

```python
data = pd.DataFrame()
for stock in stocks:
    df = yf.download(stock, start='2021-05-01', end='2025-05-01')[['Close']].rename(columns={'Close': stock})
    data = pd.concat([data, df], axis=1)

corr_matrix = data.corr(method='pearson')

print("\nMatriks Korelasi Pearson antar saham:\n")
print(corr_matrix)
```

# Analisis Distribusi

```python
from scipy.stats import skew, kurtosis

for stock in stocks:
    print(f"\n\n================================== {stock} ==================================\n")
    data = yf.download(stock, start='2021-05-01', end='2025-05-01')
    data = data[['Close']].dropna()
    data.rename(columns={'Close': 'Price'}, inplace=True)

    skewness = skew(data['Price'])
    kurt = kurtosis(data['Price'])

    print("Skewness:", round(skewness[0], 4))
    print("Kurtosis:", round(kurt[0], 4))
```

# Missing Value dan Outlier

```python
for stock in stocks:
    print(f"\n\n================================== {stock} ==================================\n")
    data = yf.download(stock, start='2021-05-01', end='2025-05-01')
    data = data[['Close']].dropna()
    data.rename(columns={'Close': 'Price'}, inplace=True)

    # ================================
    # Missing Value Check
    # ================================
    print('Jumlah Missing Value:', (data['Price'].isnull().sum()))

    # ================================
    # Outlier Check
    # ================================
    mean_price = float(data['Price'].mean())
    std_price = float(data['Price'].std())

    upper_bound = mean_price + 3 * std_price
    lower_bound = mean_price - 3 * std_price

    print("Mean Price:", mean_price)
    print("Std Dev:", std_price)
    print("Upper Bound:", upper_bound)
    print("Lower Bound:", lower_bound)

    outliers = data[(data['Price'] > upper_bound) | (data['Price'] < lower_bound)]

    plt.figure(figsize=(14, 6))
    plt.plot(data.index, data['Price'], label='Data', color='blue')

    # Garis mean, upper, lower
    plt.axhline(mean_price, color='orange', linestyle='--', label='Mean')
    plt.axhline(upper_bound, color='red', linestyle='--', label='Upper Bound (Mean+3*Std)')
    plt.axhline(lower_bound, color='green', linestyle='--', label='Lower Bound (Mean-3*Std)')
    plt.scatter(outliers.index, outliers['Price'], color='red', label='Outliers', marker='o')

    plt.title('Data Set '+ stock + ' dengan Outlier')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# Analisis PACF

```python
from statsmodels.graphics.tsaplots import plot_pacf

fig, axes = plt.subplots(3, 2, figsize=(12, 12))
axes = axes.flatten()

for i, stock in enumerate(stocks):
    data = yf.download(stock, start='2021-05-01', end='2025-05-01')
    data = data[['Close']].dropna()
    data.rename(columns={'Close': 'Price'}, inplace=True)

    plot_pacf(data['Price'], lags=40, ax=axes[i], title=f"PACF {stock}", method="ywm")

fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()

```
